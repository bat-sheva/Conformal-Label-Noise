```python
import pandas as pd
import glob
import os
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 16})
import seaborn as sns
sns.set(rc={'figure.figsize':(6,4)})
sns.set(font_scale=1.5, style='white')

directoryPath = os.path.join('results_cifar10_final/')

final_results = pd.DataFrame()
for file_name in glob.glob(directoryPath+'*.csv'):
    file = pd.read_csv(file_name, low_memory=False)
    final_results = pd.concat([final_results,file], axis=0)

final_results = final_results.drop(columns=['index'])

final_results
# final_results.to_csv('final_results_all.csv') 
```

```python
my_pal = { "clean": "b", "noisy": "g"}
ax = sns.violinplot(x='Method', y='Coverage', data=final_results, palette=my_pal, hue='Calib', split=True)
ax.set(xlabel=None)
plt.axhline(y=0.9, color='r', linestyle='--', label='nominal')
plt.grid()
plt.legend(loc='upper right')
plt.show()
fig = ax.get_figure()
fig.tight_layout()
# fig.savefig('coverage_clean_train.pdf')
```

```python
my_pal = { "clean": "b", "noisy": "g"}
ax = sns.violinplot(x='Method', y='Size', data=final_results, palette=my_pal, hue='Calib', split=True)
ax.set(xlabel=None)
plt.grid()
plt.legend(loc='upper left')
plt.show()
fig = ax.get_figure()
fig.tight_layout()
# fig.savefig('sets_clean_train.pdf')
```
