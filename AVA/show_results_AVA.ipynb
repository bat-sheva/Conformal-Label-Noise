```python
import pandas as pd
import glob
import os
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 16})
import seaborn as sns
sns.set(rc={'figure.figsize':(6,4)})
sns.set(font_scale=1.5, style='white')

directoryPath = os.path.join('results_AVA_noisy_final/')

final_results = pd.DataFrame()
for file_name in glob.glob(directoryPath+'*.csv'):
    file = pd.read_csv(file_name, low_memory=False)
    final_results = pd.concat([final_results,file], axis=0)

final_results
```

```python
my_pal = {"noisy": "g", "clean": "b"}
ax = sns.violinplot(x='Method', y='Coverage', data=final_results, palette=my_pal, hue='Calib set', split=True)
ax.set_xticklabels(['Residual magnitude', 'CQR'])
plt.axhline(y=90, color='r', linestyle='--', label='nominal')
plt.grid()
plt.legend(loc='center')
plt.show()
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('coverage_noisy_train.pdf')
```

```python
ax = sns.violinplot(x='Method', y='Length', data=final_results, palette=my_pal, hue='Calib set', split=True)
ax.set_xticklabels(['Residual magnitude', 'CQR'])
plt.grid()
plt.legend([],[], frameon=False)
# plt.legend(loc='center')
plt.show()
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('length_noisy_train.pdf')
```
